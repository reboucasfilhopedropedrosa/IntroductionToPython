## **3.1) Imagem**


> O processamento de imagem tem intuito de melhorar a qualidade,  extrair informações relevantes ou realizar tarefas específicas. Dessa forma temos alquns pré-processamento utilizados para alcançar tais objetivos, tais como **redimensionamento**, **conversão de tons**, **normalização**, **flitros**, **equalização de histograma**, **segmentação**, **remoção de ruídos**, **transformações geométricas** e entre outras. Para pré-processamento de imagens vamos usar a lib **Pillow** e a lib **OpenCV** .

### **3.1.1) Abrindo e redimensionando**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread("/content/drive/MyDrive/Trainee_Lapisco/150px-DeadTree.jpg")

# Redimensionar a imagem
novo_tamanho = (800, 600)
imagem_redimensionada = cv2.resize(imagem, novo_tamanho)

# Plotar a imagem redimensionada
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(imagem_redimensionada, cv2.COLOR_BGR2RGB))
plt.title("Imagem Redimensionada")
plt.axis("off")
plt.show()
```

### **3.1.2) Convertendo em tons**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread("/content/drive/MyDrive/Trainee_Lapisco/150px-DeadTree.jpg")

# Converter a imagem para tons de cinza
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Plotar a imagem em tons de cinza
plt.figure(figsize=(8, 6))
plt.imshow(imagem_cinza, cmap="gray")
plt.title("Imagem em Tons de Cinza")
plt.axis("off")
plt.show()
```

### **3.1.3) Normalizando**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread("/content/drive/MyDrive/Trainee_Lapisco/150px-DeadTree.jpg")

# Converter a imagem para tons de cinza
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Normalizar os valores de pixel para o intervalo [0, 1]
imagem_normalizada = cv2.normalize(imagem_cinza, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

# Plotar a imagem normalizada
plt.figure(figsize=(8, 6))
plt.imshow(imagem_normalizada, cmap="gray")
plt.title("Imagem Normalizada")
plt.axis("off")
plt.show()
```

### **3.1.3) Aplicando o filtro da média**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread("/content/drive/MyDrive/Trainee_Lapisco/150px-DeadTree.jpg")

# Aplicar filtro de média
imagem_filtrada = cv2.GaussianBlur(imagem, (5, 5), 0)

# Plotar a imagem filtrada
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(imagem_filtrada, cv2.COLOR_BGR2RGB))
plt.title("Imagem Filtrada")
plt.axis("off")
plt.show()
```

### **3.1.4) Equalização de histograma**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread("/content/drive/MyDrive/Trainee_Lapisco/150px-DeadTree.jpg")

# Converter a imagem para escala de cinza
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Aplicar filtro de média
imagem_filtrada = cv2.GaussianBlur(imagem_gray, (5, 5), 0)

# Equalização de histograma
imagem_equalizada = cv2.equalizeHist(imagem_filtrada)

# Plotar a imagem equalizada
plt.figure(figsize=(8, 6))
plt.imshow(imagem_equalizada, cmap="gray")
plt.title("Imagem Equalizada")
plt.axis("off")
plt.show()
```

## **3.2) Sinais**

> O processamento de sinais permite analisar e manipular dados em formato de sinais, como áudio, séries temporais e outros, para extrair informações relevantes e realizar diversas tarefas.  Dessa forma temos alquns pré-processamento utilizados para alcançar tais objetivos, tais como **Amostragem e Reconstrução de Sinais**, **Transformada de Fourier**, **Filtragem de Sinais (Passa-baixa):**, **Análise Espectral:** e entre outras. Utilizaremos para análise a lib numpy.

### **3.2.1) Amostragem e Reconstrução de Sinais**

```python
import numpy as np
import matplotlib.pyplot as plt

# Sinal contínuo
t = np.linspace(0, 1, 1000)  # Vetor de tempo de 0 a 1 segundos
frequencia_sinal = 5  # Frequência do sinal
sinal_continuo = np.sin(2 * np.pi * frequencia_sinal * t)

# Amostragem
frequencia_amostragem = 100  # Frequência de amostragem em Hz
amostras = np.arange(0, 1, 1 / frequencia_amostragem)
sinal_amostrado = np.sin(2 * np.pi * frequencia_sinal * amostras)

# Reconstrução
sinal_reconstruido = np.interp(t, amostras, sinal_amostrado)

# Plotar os resultados
plt.figure(figsize=(10, 6))
plt.plot(t, sinal_continuo, label='Sinal Contínuo')
plt.stem(amostras, sinal_amostrado, 'r', markerfmt='ro', label='Sinal Amostrado')
plt.plot(t, sinal_reconstruido, 'g', label='Sinal Reconstruído')
plt.legend()
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Amostragem e Reconstrução de Sinais')
plt.grid(True)
plt.show()
```

### **3.2.2) Transformada de Fourier**

```python
import numpy as np
import matplotlib.pyplot as plt

# Sinal contínuo
t = np.linspace(0, 1, 1000)  # Vetor de tempo de 0 a 1 segundos
frequencia_sinal = 5  # Frequência do sinal
sinal_continuo = np.sin(2 * np.pi * frequencia_sinal * t)

# Transformada de Fourier
transformada = np.fft.fft(sinal_continuo)
frequencias = np.fft.fftfreq(len(t), t[1] - t[0])

# Plotar a magnitude da Transformada de Fourier
plt.figure(figsize=(8, 6))
plt.plot(frequencias, np.abs(transformada))
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Transformada de Fourier')
plt.grid(True)
plt.show()
```

### **3.2.3) Filtragem de Sinais (Passa-baixa)**

```python
import numpy as np
import matplotlib.pyplot as plt

# Sinal contínuo
t = np.linspace(0, 1, 1000)  # Vetor de tempo de 0 a 1 segundos
frequencia_sinal = 10  # Frequência do sinal
sinal_continuo = np.sin(2 * np.pi * frequencia_sinal * t)

# Adicionar ruído ao sinal
ruido = 0.5 * np.random.randn(len(t))
sinal_com_ruido = sinal_continuo + ruido

# Filtragem Passa-baixa
frequencia_corte = 15  # Frequência de corte do filtro
transformada = np.fft.fft(sinal_com_ruido)
transformada[frequencias > frequencia_corte] = 0
sinal_filtrado = np.fft.ifft(transformada)

# Plotar os resultados
plt.figure(figsize=(10, 6))
plt.plot(t, sinal_continuo, label='Sinal Original')
plt.plot(t, sinal_com_ruido, 'r', label='Sinal com Ruído')
plt.plot(t, sinal_filtrado, 'g', label='Sinal Filtrado')
plt.legend()
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Filtragem Passa-baixa')
plt.grid(True)
plt.show()
```

### **3.2.4) Análise Espectral**

```python
import numpy as np
import matplotlib.pyplot as plt

# Sinal contínuo
t = np.linspace(0, 1, 1000)  # Vetor de tempo de 0 a 1 segundos
frequencia_sinal = 5  # Frequência do sinal
sinal_continuo = np.sin(2 * np.pi * frequencia_sinal * t)

# Transformada de Fourier e análise espectral
transformada = np.fft.fft(sinal_continuo)
frequencias = np.fft.fftfreq(len(t), t[1] - t[0])
espectro = np.abs(transformada)

# Plotar o espectro de frequência
plt.figure(figsize=(8, 6))
plt.plot(frequencias, espectro)
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Análise Espectral')
plt.grid(True)
plt.show()
```
