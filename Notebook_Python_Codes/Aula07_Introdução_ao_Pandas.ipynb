> Pandas tem dois tipos principais de estruturas para organiza√ß√£o dos dados: Series (1-Dimens√£o) e Dataframe (2-Dimens√µes).

#### Series

```python
# Pandas Series

labels = ['linha 1', 'linha 2', 'linha 3']  # lista strings
my_list = [10, 20, 30] # lista num√©rica
array = np.array([1.1, 1.2, 1.3])  # numpy array
dic = {'x' : 10, 'y' : 20, 'z' : 30}  # dicion√°rio
t = (2, 4, 5, 7) # tupla

serie_1 = pd.Series(array)
serie_1
```

Atente acima ao tipo dos dados da Serie

```python
# Acessar os valores de uma serie

serie_1[2] # indice numerico
# serie_1['x'] # indice string
```

```python
# setar index de uma serie utilizando os argumentos da fun√ßao

serie_2 = pd.Series(data=labels, index=my_list)
serie_2
```

#### Dataframe

```python
from numpy.random import rand

matriz = np.random.rand(5, 4)  # gerando uma matriz (array 2-D) com n√∫meros aleat√≥rios
print(matriz)
```

```python
# Dataframe
col = ["col 1", "col 2", "col 3", "col 4"]
dfnum = pd.DataFrame(data=matriz, columns=col)
dfnum
```

```python
# Dataframe a partir de um dicionario

dicionario = {'nome' : ['maria', 'ana','joao'],
              'idade': [10, 12, 14],
              'altura' : [1.75, 1.67, 1.45]}

dfpessoa = pd.DataFrame(dicionario)

dfpessoa
```

##### **Selecionar linhas e colunas**

```python
# Selecionar colunas de um DataFrame

dfpessoa['idade'] # ao selecionar uma coluna, o resultado √© uma Serie
idades = dfpessoa['idade']
type(idades)
```

```python
# Selecionar linhas e colunas pelos labels

display(dfnum.loc[:,'col 2':'col 3'])
display(dfnum.loc[1:2,'col 1'])
```

```python
# selecionar linhas pelo indice numerico

display(dfnum.iloc[:,1])
display(dfnum.iloc[0:1,1:2])
```

```python
# sele√ßao condicional
# operadores relacionais > < >= <= == != (True ou False)

display(dfnum[dfnum['col 1'] < 0.40])
display(dfnum)
```

##### **Adicionar e remover colunas**

```python
# adicionar novas colunas
dfpessoa['nova_col'] = [20,30,70]
dfpessoa
```

```python
# remover colunas
dfpessoa.drop('nova_col', axis=1, inplace=True) # axis=1 para colunas e axis=0 para linhas
dfpessoa
```

##### **Valores faltantes**

Pode acontecer de um dataset ter em suas linhas e colunas valores faltantes, c√©lulas da tabela em que n√£o h√° valor algum. Geralmente esses dados (a aus√™ncia deles, no caso) **n√£o s√£o importantes** para an√°lises de dados e para treinamento de algoritmos, precisando portanto serem tratados.

> Os valores faltantes em um .csv comumente v√£o assumir o tipo **'np.nan'**, um nan (not a number) √© uma constante que representar√° um valor que n√£o temos ou que n√£o foi definido. (outras representa√ß√µes s√£o NaN ou NAN)

```python
# criaremos um novo dataframe para incluirmos nele NaNs

dicio = {'A' : [11, 52, 33, np.nan], 'B' : [15, 40, np.nan, np.nan], 'C' : [10, 12, 43, 77]}

df_nan = pd.DataFrame(data = dicio)
df_nan
```

```python
# o m√©todo info nos d√° informa√ß√µes sobre nosso dataframe
df_nan.info()
```

```python
df_nan.isnull() #identificar os registros faltantes
```

```python
# um artificio muito utilizado para contar o numero de NaNs no dataframe
df_nan.isnull().sum()
```

> Uma maneira de tratar os valores faltantes √© remover as linhas que cont√©m por ao menos um:

```python
df_nan.dropna() # remover as linhas que cont√©m ao menos um registro faltante
```

S√≥ que, dependendo do dataset, poder√° n√£o dispor de muitas linhas e dados s√£o muito valiosos! quanto mais tiver, melhor poder√° ser a an√°lise, tomada de decis√£o e o treinamento dos algoritmos.

```python
df_nan.fillna(value=0) # substituir por algum valor padr√£o em todo dataframe
```

```python
df_nan.fillna(value={'A': 1, 'B': np.mean(df_nan['B'])}, inplace = True) # substituir em colunas indicadas
df_nan
```

##### **Juntar dataframes**

```python
# concatenando dataframes

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                     index=['a', 'b', 'c', 'd'])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                     index=['e', 'f', 'g', 'h'])
display(df1)
display(df2)
```

```python
df3 = pd.concat([df1, df2])
display(df3)
```

**M√©todo Join**

```python
df4 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                  )
df5 = pd.DataFrame({'E': ['A4', 'A5', 'A6', 'A7'],
                    'F': ['B4', 'B5', 'B6', 'B7'],
                    'G': ['C4', 'C5', 'C6', 'C7'],
                    'H': ['D4', 'D5', 'D6', 'D7']},
                  )

df6 = df4.join(df5)
display(df6)
```
