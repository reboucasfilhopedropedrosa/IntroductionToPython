## **2) Manipulação de arquivo**

> O arquivo é  um **conjunto de dados** digitalizados que pode ser gravado em um dispositivo de armazenamento e tratado como ente único.

### **2.1) Criando e abrindo arquivos .txt**

> Para criar e abrir podemos utilizaro metodo **open ( )** , dessa forma iremos abrir com um parametro determinado, de acordo com o modo de uso


> Para escrever dentro do arquivo vamos utilizar o método **write ( )**, que recebe uma string como parâmetro e a insere no arquivo.

Captura de tela 2023-07-11 082639.png

```python
nome_arquivo = 'arquivo_escrita.txt'
conteudo_para_anexar = 'Trainee de Programação Básica em Python - Lapisco'

with open(nome_arquivo, 'a') as arquivo:
    arquivo.write(conteudo_para_anexar)
```

```python
nome_arquivo = 'arquivo_escrita.txt'
conteudo_para_escrever = 'Trainee de Programação Básica em Python - Lapisco - 2ª aula'

with open(nome_arquivo, 'w') as arquivo:
    arquivo.write(conteudo_para_escrever)
```

###**2.2) Lendo arquivos**

> Para fazer a leitura os dados que os arquivos possuem , vamos utilizar dois métodos, primeiro **readline ( )**, que lê a quantidade N de caracteres da primeira linha passadas com parâmetro. E o segundo é o **readlines ( )** que retorna todas linhas.

```python
nome_arquivo = 'arquivo_leitura1.txt'
conteudo_para_anexar = 'Trainee de Programação Básica em Python - Lapisco'

with open(nome_arquivo, 'a') as arquivo:
    arquivo.write(conteudo_para_anexar)

with open(nome_arquivo, 'r') as arquivo:
    conteudo_atualizado = arquivo.readline(10)

print(conteudo_atualizado)
```

```python
nome_arquivo = 'arquivo_leitura2.txt'
conteudo_para_anexar = 'Trainee de Programação Básica em Python - Lapisco'

with open(nome_arquivo, 'a') as arquivo:
    arquivo.write(conteudo_para_anexar)

with open(nome_arquivo, 'r') as arquivo:
    conteudo_atualizado = arquivo.readlines()

print(conteudo_atualizado)
```

### **2.3.4) Manipulação de arquivos**

> Para a munipulação de arquivos temos vários comandos, mas vou listar os mais usados :
 - `.head()`
 - `.tail()`
 - `.describe()`
 - `.shape()`
 - `.dtypes()`
 - `.loc()`
 - `.iloc()`
 - `.sort_values()`
 - `.isnul()`

> No método **head ( )** como seguimos ver uma quantidade N de dados de uma tebla, contando de cima para baixo

```python
# Exemplo com head
new_df.head(1)
```

> No método **tail ( )** como seguimos ver uma quantidade N de dados de uma tebla, contando de baixo para cima

```python
# Exemplo com tail
new_df.tead(1)
```

> No método **decribe ( )** ele decreve todos os dados presentes, criando uma linha para count (conta a quantidade de dados), mean (dá media dos valores), std(dá o desvio padrão), min (dá o minimo valor dos dados), 25%, 50%, 75% (representam os valores em repectivas porcentagens), max( máximo valor presente)

```python
# Exemplo com describe
new_df.describe()
```

> No método **shape ( )**  mostra o tamenho da nossa tabela em x e y

```python
# Exemplo com shape
new_df.shape
```

> No método **dtypes ( )** mostra o tipo de objeto que a coluna recebe

```python
# Exemplo com dtypes
new_df.dtypes
```

> No método **loc ( )** permite acessar o DataFrame a partir do nome, acessando pelo índice da linha ou coluna.

```python
# Exemplo com loc
new_df.loc[(new_df['Nome'] == 'João')]
```

> O metodo **iloc ( )**  permite você acessar um DataFrame a partir do índice da linha e da coluna. Recebendo valor inteiro(int) para acessar a linha ou coluna.

```python
# Exemplo com iloc
new_table.iloc[1:2]
```

> O metodo **sort_values ( )**  ordena os registros com base em uma ou mais colunas do seu DataFrame, permitindo escolhar a forma de ordenção.

```python
# Exemplo com sort_values
new_table.sort_values(by=['Nome'])
```

> O metodo **isnull ( )**  usados para verificar e gerenciar valores NULL em um dataframe.

```python
# Exemplo com isnull
new_table.isnull()
```
