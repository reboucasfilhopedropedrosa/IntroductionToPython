### Tarefas típicas de Machine Learning

- Classificação: A que realizaremos neste Trainee será esta. A tarefa de classificar algo é fundamental para a resolução de muitas problemáticas do mundo real, o ser humano classifica coisas de maneira natural e muitas vezes sem perceber. Nós utilizaremos um algoritmo para implementar esta tarefa.

- Regressão: A tarefa de por exemplo prever um valor de um carro ou prever o valor de uma casa é uma tarefa de regressão. Constantemente o ser humano também se vale de tentar prever valores numéricos com base em uma bagagem que possui.


### Um desafio em Machine Learning: Dados insuficientes



> Certamente será um desafio para você treinar um bom modelo de Inteligência Artificial se não tiver uma boa quantidade de dados, uma grande base com dados bem detalhados. Poucos dados pode significar que o algoritmo não aprenda o suficiente para poder responder com uma precisão satisfatória à sua tarefa.

### Outro desafio: A qualidade dos dados

> Chamamos de OUTLIER um dado em que foge da realidade de uma coluna de dados por exemplo. Ter outliers em um dataset é muito perigoso pois ele será utilizado no aprendizado do algoritmo e pode interferir na sua capacidade de generalizar, passando portanto por um sobrejuste (OVERFITTING). O que, para uma problemática real, é muito ruim dado que ele sabe muito bem classificar ou predizer os dados de treinamento mas se sairá muito mal realizando a mesma tarefa em dados que nunca viu.

## Implementando Machine Learning no nosso Dataset

### Treinamento, Teste e Validação.

Geralmente se divide os dados que temos afim de uma parte seja usada para ensinar o algoritmo, outra para validar como o algoritmo tá aprendendo, se ele tá aprendendo direito e outra parte com dados que ele NUNCA viu (não usados na parte de treinamento) para testar e ver como o algoritmo já treinado se sai em dados que nunca viu.

- Treinamento: É a maior parcela da divisão, ficando sempre por volta de 70 a 80% do dataset inteiro para treinar e ensinar os padrões dos dados afim de que ele consiga ter uma "boa bagagem" para exercer a sua tarefa.

- Validação: No nosso trainee não utilizaremos essa parte mas deve ser realizada por todo aquele que já tem um domínio iniciante de Machine Learning. Por a divisão mais comum ser 80 para treino, 10 para validação e 10 para teste, utilizaríamos no caso 10% dos dados para validar o treinamento e ver como o modelo aprende ao longo do treinamento.

- Teste: É IMPORTANTE EXPLICITAR QUE nesta etapa o algoritmo NÃO pode ser testado com dados que ele viu em treino. Deve-se portanto sempre ter esse cuidado para que os dados de treino não estejam sendo usados para teste afim de não fraudar o teste e o algoritmo não aprender bem ou pior, só aprender aquele contexto, sofrer OVERFITTING.

#### Exemplo de Implementação e atividade para casa

Temos um dataset que terá, após todo o pré-processamento e a EDA, aproximadamente 10-12 colunas. Dessas 10-12 colunas, UMA é a coluna que queremos classificar, no nosso caso a coluna "Survived" e a outras são as colunas que usaremos para ensinar o padrão para o algoritmo. Logo chamaremos a coluna que queremos classificar de y e todas as outras de X. Com base nessa divisão, dividiremos o dataset em 80% para treino e 20% para teste. Após isso, instanciamos o algoritmo e o treinamos com a parte do conjunto para treino. Após o treinamento o testamos e recolhemos suas predições, afim de que se utilize elas para calcularmos métricas para avaliar como o modelo se saiu no teste.
